**The laureate who won more than one nobel prize**

```python
q='''
SELECT full_name, 
       COUNT(year) AS prize_count 
FROM   nobel 
GROUP  BY full_name 
HAVING prize_count > 1 
ORDER  BY prize_count DESC 
'''
ex_df(q)
```

[Comité international de la Croix Rouge](https://fr.wikipedia.org/wiki/Comit%C3%A9_international_de_la_Croix-Rouge) (International Committee of the Red Cross) has won three nobel.

But wikipedia page of [Jack W. Szostak](https://en.wikipedia.org/wiki/Jack_W._Szostak) says he won nobel once in 2009 and not three times. So do we have duplicates in our table.

### Duplicate rows in table

```python
q="""
-- for each name from inner query print info 
SELECT year, 
       category, 
       laureate_id, 
       full_name, 
       birth_date 
FROM   nobel 
WHERE  full_name IN ( 
                    -- gives list of names for which duplicates are present 
                    SELECT full_name 
                     FROM   nobel 
                     GROUP  BY year, 
                               full_name 
                     HAVING Count(*) > 1) 
"""
ex_df(q)
```

Yes we have many duplicate entries.
